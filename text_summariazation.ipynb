```python
#2. prompt template text summariazation
from langchain.chains import LLMChain
from langchain import PromptTemplate

generictemplate= """AIMessagewrite a summary of the following text:
text:{text}
Translate the precise summary to {language}
""" 

prompt=PromptTemplate(
    input_variables=['text','language'],
    template=generictemplate
)
prompt
```

```python
complete_prompt=prompt.format(text=text,language="french")
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain=LLMChain(llm=llm,prompt=prompt)
summary=llm_chain.run({'text':text,'language':'hindi'})
summary
```

```python
#stuffDocumentChain text summariazation

from langchain_community.document_loaders import PyPDFLoader

loader=PyPDFLoader("apjspeech.pdf")
docs=loader.load_and_split()
docs
```

```python
template=""" 
write a consise and a short summary of the speech,
speech:{text}
"""

prompt=PromptTemplate(input_variable=['text'],
                      template=template)
```

```python
from langchain.chains.summarize import load_summarize_chain
```

```python
chain=load_summarize_chain(llm,chain_type='stuff',prompt=prompt,verbose=True)
output_summary=chain.run(docs)
output_summary
```

```python
#map reduce summaraization
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
docs
```

```python
final_documents=RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=100).split_documents(docs)
final_documents
```

```python
len(final_documents)
```

```python
prompt
```

```python
from langchain import PromptTemplate
chunk_prompt=""" 
please summarazie the below apeech:
speech:'{text}' 
summary:
""" 

map_prompt_template=PromptTemplate(input_variables=['text'],
                                   template=chunk_prompt)
```

```python
final_prompt = ''' 
provide the final summary of the entire speech with these important points.
add a motivation title, start the precise summary with an introduction and provide the summary in number of points for the speech.
Speech:{text}
'''
final_prompt_template=PromptTemplate(input_variables=['text'],template=final_prompt)
final_prompt_template
```

```python
from langchain.chains.summarize import load_summarize_chain
summary_chain=load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt=map_prompt_template,
    combine_prompt=final_prompt_template,
    verbose=True
)

output=summary_chain.run(final_documents)
```

```python
#refine chain summariazation

chain=load_summarize_chain(
    llm=llm,
    chain_type="refine",
    verbose=True
)

output_summary=chain.run(final_documents)
```
